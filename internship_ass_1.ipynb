```python
import pandas as pd

import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression

from sklearn.metrics import mean_squared_error
```

```python
df = pd.DataFrame(
    {

'x': [140, 155, 159, 179, 182,200,212],
'y': [60, 62,67,70,71,72,75]
})
```

```python
model= LinearRegression()
model.fit(df[['x']], df['y'])
#y_predicted: These are the predicted y-values obtained from a linear regr

y_predicted = model.predict(df[['x']])
print(y_predicted)
```

```python
mse = mean_squared_error(df[ 'y'], y_predicted)
#error=Y-Predicted
# Print the calculated mean squared error
print("Mean Squared Error:", mse)
```

```python
plt.scatter(df[ 'x'], df['y'], label='Actual Data')
plt.plot(df['x'], y_predicted, color='red', label='Linear Regression Line')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression')
plt.legend()
plt.show()
```
